```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("student.csv")
print(df.head())
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="grade", data=df)
plt.xlabel('Grades')
plt.title("Box Plot for Grades")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.violinplot(x="grade", data=df)
plt.xlabel('Grades')
plt.title("Violin Plot for Grades")
plt.show()
```
