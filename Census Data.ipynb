```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

dt= pd.read_csv('nyc_census_tracts.csv')
print(dt.head())
```

```python
plt.figure(figsize=(8,6))
plt.hist(dt['Income'], bins=20, color='skyblue')
plt.xlabel('Income')
plt.ylabel('Frequency')
plt.title('Income Histogram')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.scatterplot(x='education', y='Income', data=dt, color='purple')
plt.xlabel('Education')
plt.ylabel('Income')
plt.title('Education and Income')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.bar(dt['education'], dt['Income'], color='red')
plt.xlabel('Education')
plt.ylabel('Income')
plt.title('Education and Income')
plt.xticks(rotation=90)  # Rotate x-axis labels for better visibility if needed
plt.show()
```
